```python
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans,DBSCAN
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.preprocessing import scale
from sklearn import cluster
from sklearn import metrics
```

```python
title = ["Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols"
        ,"Flavanoids","Nonflavanoid phenols","Proanthocyanins",
        "Color intensity","Hue","OD280/OD315 of diluted wines",
        "Proline"]
file = pd.read_csv("./wine.data",header=None)
file.shape
```

```python
data = file.iloc[:,1:]
data.columns = title
label = file[0]
```

```python
# sample = np.linspace(0.5,10,96)
dic = {"eps":[],"samples":[],"num":[]}
for i in range(1,100):
    for j in range(1,100):
        dbscan = DBSCAN(eps=i,min_samples=j)
        res = dbscan.fit_predict(data)
        Max = res.max()
        if  Max== 1 and len(res[res==0])>20:
            print("[eps:{}][min_sample:{}]".format(i,j))
            print(res)
            dic["eps"].append(i)
            dic["samples"].append(j)
            dic["num"].append(len(list(set(res))))
#eps:52][min_sample:27]
```

```python
# dic
pd.DataFrame(data=dic).to_csv("r.csv")
```

```python

dbscan = DBSCAN(eps=98,min_samples=21)
result = dbscan.fit_predict(data)
y_pred2 = result+2
y_pred2

f1_micro = metrics.f1_score(label,y_pred2,average="micro",)
f1_macro = metrics.f1_score(label,y_pred2,average="macro")
print(f1_micro)
print(f1_macro)
metrics.silhouette_score(data,y_pred2,metric='euclidean')
```
